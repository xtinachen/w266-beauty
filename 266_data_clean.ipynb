# Data Clean/Wrangle

```
# One function to clean
def clean(text):
  text = text.lower()                                                    #lowercase string
  text = remove_punctuation(text)                                             #remove punctuations
  text = ' '.join([word for word in text.split() if word not in (stop_words)])   #remove stopwords
  # text = ' '.join([word for word in text.split() if not word.isdigit()])         #remove numbers
  text = re.sub('[0-9]+', '', text) # remove number
  # stemming below
  token_text = nltk.word_tokenize(text)
  stem_words = [stemmer.stem(w) for w in token_text] # use snowball?
  stem_words = ' '.join(stem_words)
  # lemmatization below
  # text = find_lemmas(stem_words)

  return stem_words
```
